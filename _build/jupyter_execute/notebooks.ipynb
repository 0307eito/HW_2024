## 全台灣釣具行分佈



```python
import requests
import json
import folium
import overpass
```

```python
overpass_url = "http://overpass-api.de/api/interpreter"

# 您可以根據需求修改這裡的查詢條件
overpass_query = """
[out:json];
node
  ["shop"="fishing"]
  (21.8,119.5,25.4,122.0);
out;
"""

response = requests.get(overpass_url, 
                        params={'data': overpass_query})
data = response.json()
```

```python
m = folium.Map(location=[23.7, 121.0], zoom_start=8)

# 遍歷 Overpass API 的回應資料，並將每個點標記在地圖上
for element in data['elements']:
    lat = element['lat']
    lon = element['lon']
    name = element.get('tags', {}).get('name', 'Fishing Port')
    
    folium.Marker(
        location=[lat, lon],
        popup=name,
        icon=folium.Icon(icon='info-sign')
    ).add_to(m)

# 顯示地圖
m
```
